## Pie Plot (helper)

```python
labels
```

```python
title = plt.title('...')
title.set_ha("left")
plt.gca().axis("equal")
pie = plt.pie(dfmeans['price'], startangle=0)

plt.legend(pie[0], labels, bbox_to_anchor=(1,0.5), loc="center right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.45)
```
